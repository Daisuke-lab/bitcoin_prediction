```python
import sympy as sym
from sympy.plotting import plot
sym.init_printing(use_unicode=True)
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats
import math
import math
%matplotlib inline
```

```python
mu = 65869.00442643142
sigma = 40138.47690132205
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
fig, ax = plt.subplots()
ax.plot(x, stats.norm.pdf(x, mu, sigma))
ax.scatter([60697.3], [0])
plt.show()
```

```python
f = sym.Function('f')
a, b, c, x, y = sym.symbols("a b c x y")
```

```python
eq = sym.Eq(-(f(x).diff(x,2)/f(x).diff(x,1)), 0.18)
eq
```

```python
ans = sym.dsolve(eq)
print(ans)
```

```python
ans = sym.dsolve(eq, ics={f(0):0})
print(ans)
```

```python
eq = sym.Eq(-(f(x).diff(x,2)/f(x).diff(x,1)), 0.18)
ans = sym.dsolve(eq)
print(ans)
ans = sym.dsolve(eq, ics={f(0):0, f(65869.00442643142):40000})
print(ans)
plot(ans.rhs,x, (x, 0, 70000))
```
