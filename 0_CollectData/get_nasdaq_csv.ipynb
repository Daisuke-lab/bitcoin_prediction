```python
#https://data.nasdaq.com/data/BCHAIN-blockchain
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import gzip
import shutil
import glob
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options  # クロームドライバー
from selenium.webdriver.common.action_chains import ActionChains
from selenium import webdriver
import chromedriver_binary
import datetime
import os
from time import sleep
```

```python
class GetCsv():
    def __init__(self):
        self.df_production = None
        self.email = "kikuchidaisuke.zr@gmail.com"
        self.password = "3153Zr0314"
        self.start_chrome_driver()
        
    def start_chrome_driver(self):
        self.driver = webdriver.Chrome()
        self.main_page ='https://data.nasdaq.com/data/BCHAIN-blockchain?page=1'
        self.driver.get("https://data.nasdaq.com/login")
        self.driver.find_element_by_xpath('//*[@data-test-text-field-id="email"]').find_element_by_xpath('label/input').send_keys(self.email)
        self.driver.find_element_by_xpath('//*[@data-test-text-field-id="password"]').find_element_by_xpath('label/input').send_keys(self.password)
        self.driver.find_element_by_xpath('//*[@data-test-progress-button="login"]').click()
        sleep(5)
        
        
    def wait_unitl_completing_download(self):
        csv_files = []
        while len(csv_files) == 0:
            csv_files = glob.glob(r"C:\Users\daisu\Downloads\BCHAIN*.csv")
            sleep(3)
        return csv_files[0]
    
    def concat_csv(self, csv):
        df = pd.read_csv(csv)
        file_name = csv.split('\\')[-1].split('.')[0]
        df = df.rename(columns={'Value': file_name})
        if self.df_production is None:
            self.df_production = df
        else:
            self.df_production = pd.merge(self.df_production, df, how="left", on=["Date"])
        
    def main(self):
        try:
            pages = [1,2,3,4]
            for page in pages:
                self.main_page = f"https://data.nasdaq.com/data/BCHAIN-blockchain?page={page}"
                self.driver.get(self.main_page)
                sleep(2)
                length = len(self.driver.find_elements_by_xpath('//a[@data-test-product-link="expand"]'))
                for index in range(length):
                    expand = self.driver.find_elements_by_xpath('//a[@data-test-product-link="expand"]')[index]
                    expand.click()
                    sleep(5)
                    self.driver.find_element_by_xpath('//*[text()= "Download"]').click()
                    csv_link = self.driver.find_element_by_xpath('//a[@data-test-dataset-download="csv"]')
                    link = csv_link.get_attribute("href")
                    self.driver.get(link)
                    csv_file = self.wait_unitl_completing_download()
                    self.concat_csv(csv_file)
                    os.remove(csv_file)
                    self.driver.get(self.main_page)
                    sleep(5)
                self.df_production.to_csv(f"nasdaq{page}.csv")
        except Exception as e:
            print(e)
        
        finally:
            self.driver.quit()
            self.df_production.to_csv('nasdaq.csv')
            
            
```

```python
get_csv = GetCsv()
get_csv.main()
```
